* Create an overview table that summarises key metrics about each school, including:

* Create a dataframe to hold the above results

```python
 # Get the unique school names from the merged dataset
school_names = school_data_complete['school_name'].unique()
```

```python
# Calculate the total number of students for each school
total_students = school_data_complete.groupby('school_name')['Student ID'].count()
```

```python
# Calculate the school type for each school
school_types = school_data.set_index('school_name')['type']
```

```python

# Calculate the total school budget
total_budget = school_data.groupby('school_name')['budget'].first()
```

```python
# Calculate the per student budget
per_student_budget = total_budget / total_students
```

```python

# Calculate the average math score for each school
average_math_score = school_data_complete.groupby('school_name')['maths_score'].mean()
```

```python

# Calculate the average reading score for each school
average_reading_score = school_data_complete.groupby('school_name')['reading_score'].mean()
```

```python

# Calculate the percentage of students passing math for each school
passing_math_percentage = (school_data_complete[school_data_complete['maths_score'] >= 50].groupby('school_name')['Student ID'].count() / total_students) * 100
```

```python
# Calculate the percentage of students passing reading for each school
passing_reading_percentage = (school_data_complete[school_data_complete['reading_score'] >= 50].groupby('school_name')['Student ID'].count() / total_students) * 100
```

```python
# Calculate the percentage of students passing both math and reading for each school
passing_overall_percentage = (school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)].groupby('school_name')['Student ID'].count() / total_students) * 100
```

```python


# Create a DataFrame with the key metrics
metrics_df = pd.DataFrame({
    'School Name': total_students.index,
    'School Type': school_types.values,
    'Total Students': total_students.values,
    'Total School Budget': total_budget.values,
    'Per Student Budget': per_student_budget.values,
    'Average Math Score': average_math_score.values,
    'Average Reading Score': average_reading_score.values,
    '% Passing Math': passing_math_percentage.values,
    '% Passing Reading': passing_reading_percentage.values,
    '% Overall Passing': passing_overall_percentage.values
})

# Display the DataFrame
metrics_df
```

```python
# Define the spending bins and group names
spending_bins = [0, 585, 630, 645, 675]
group_names = ['<$585', '$585-630', '$630-645', '$645-675']
```

```python
# Categorize schools into spending ranges
school_data_complete['Spending Ranges (Per Student)'] = pd.cut(school_data_complete['budget'] / school_data_complete['size'], spending_bins, labels=group_names)
```

```python
# Calculate average math scores by spending range
average_math_scores = school_data_complete.groupby('Spending Ranges (Per Student)')['maths_score'].mean()
```

```python
# Calculate percentage passing math by spending range
passing_math_percentages = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('Spending Ranges (Per Student)')['Student ID'].count() / school_data_complete.groupby('Spending Ranges (Per Student)')['Student ID'].count() * 100
```

```python
# Calculate percentage passing reading by spending range
passing_reading_percentages = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('Spending Ranges (Per Student)')['Student ID'].count() / school_data_complete.groupby('Spending Ranges (Per Student)')['Student ID'].count() * 100
```
